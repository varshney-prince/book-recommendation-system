## **Popularity Based Recommendation System**

```python
## we will show highest average rating in the output

rating_with_name=rating.merge(book,on='ISBN')
```

```python
rating_with_name
```

```python
num_rating_df=rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_rating'},inplace=True)
num_rating_df
```

```python
avg_rating_df=rating_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_rating']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df
```

```python
popular_df=popular_df.merge(book,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_rating','avg_rating']]
```

```python
popular_df
```

##  **Collaborative Recommendation System**

```python
x=rating_with_name.groupby('User-ID').count()['Book-Rating'] >200
pade_likhe_user=x[x].index
```

```python
filtered_rating=rating_with_name[rating_with_name['User-ID'].isin(pade_likhe_user)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt= final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores.shape
df_cd = pd.merge(book, avg_rating_df, how='inner')
```

```python
df_cd.head()
```

```python
def recommend(book_name):
  # print((pt.index==book_name)[0][0])
    index=np.where(pt.index==book_name)[0][0]
    # print(index)
    similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    print(similar_items)
    data = []
    for i in similar_items:
        item = []
        temp_df = df_cd[df_cd['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['avg_rating'].values.round(2)))
        
        data.append(item)
  
 
    return data
```

```python
recommend('1984')
```
